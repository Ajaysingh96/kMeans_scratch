```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
import pandas as pd
```

```python
from sklearn.datasets import make_blobs
```

```python
# Create Data
data = make_blobs(n_samples=200, n_features=2, 
                           centers=3, cluster_std=1.8,random_state=101)
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
centroids = make_blobs(n_samples=3, n_features=2, 
                           centers=4, cluster_std=1.8,random_state=101)[0]
```

```python
for i in range(30):
    #print(centroids)
    dis=cdist(data[0],centroids,'euclidean')
    #print(dis.shape)
    label=np.argmin(dis,axis=1)
    hot = np.array(pd.get_dummies(label))
    #print(hot.shape)
    sum_points=np.dot(data[0].T,hot)
    #print(sum_points)
    new_centroids=(sum_points/np.sum(hot,axis=0)).T
    centroids=new_centroids

    
    
    
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=label,cmap='rainbow')
plt.scatter(centroids[:,0],centroids[:,1])
```
